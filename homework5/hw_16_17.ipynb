```python
print ("hello")
import numpy as np
import pandas as pd
from sklearn import svm
import matplotlib.pyplot as plt
```

```python
train=pd.read_table("features.train.txt",delim_whitespace=True,header=None)
train_x=train.iloc[:,1:3]
train_y=train.iloc[:,0]
train_y=(train_y==8)*2-1
test=pd.read_table("features.test.txt",delim_whitespace=True,header=None)
test_x=test.iloc[:,1:3]
test_y=test.iloc[:,0]
test_y=(test_y==8)*2-1
```

```python
C=[-6,-4,-2,0,2]
sigmaa=[]
Ein=[]
for c in C:
    clf=svm.SVC(C=pow(10,c),kernel='poly',degree=2,gamma=1,coef0=1)
    clf.fit(train_x,train_y)
    alpha=abs(clf.dual_coef_)
    sigmaa.append(np.sum(alpha))
    Ein.append(clf.score(train_x,train_y))
print (sigmaa)
print (Ein)
plt.plot(C,Ein)
plt.show()
plt.plot(C,sigmaa)
plt.show()
```
