```python
import numpy as np
import pandas as pd
from sklearn import svm
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
train=pd.read_table("features.train.txt",delim_whitespace=True,header=None)
train_x=train.iloc[:,1:3]
train_y=train.iloc[:,0]
train_y=(train_y==0)*2-1
test=pd.read_table("features.test.txt",delim_whitespace=True,header=None)
test_x=test.iloc[:,1:3]
test_y=test.iloc[:,0]
test_y=(test_y==0)*2-1
#train_x[train_y==1]
```

```python
Gammalist=[]
for i in range(100):
    print (i)
    tra_x,val_x,tra_y,val_y=train_test_split(train_x,train_y,test_size=1000)
    Gamma=[1,2,3,4]
    final_gamma=0
    maxscore=0
    for gamma in Gamma:
        clf=svm.SVC(C=pow(10,0.1),kernel='rbf',gamma=gamma)
        clf.fit(tra_x,tra_y)
        score=clf.score(val_x,val_y)   
        print (score)
        if(score>maxscore):
            final_gamma=gamma
            maxscore=score
    print ("\n")
    Gammalist.append(final_gamma)
Gammalist
```

```python
plt.hist(Gammalist)
plt.show()
```
