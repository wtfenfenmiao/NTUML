```python
def kernel(x1,x2,gama):
    return np.exp(-gama*np.dot((x1-x2),(x1-x2).T))
```

```python
def getK(X,gama):
    K=np.zeros((len(X),len(X)))
    for i in range(len(X)):
        for j in range(i,len(X)):   
            K[i][j]=K[j][i]=kernel(X.iloc[i,0:-1],X.iloc[j,0:-1],gama)
    return K
```

```python
def sign(X):
    return (X>0)*2-1
```

```python
def predict(beta,train,test_one):
#     print (beta)
#     print (beta.shape)
#     print ((train-test_one)*(train-test_one))
#     print (np.sum((train-test_one)*(train-test_one),axis=1))
#     print (-gama*np.sum((train-test_one)*(train-test_one),axis=1))
#     print (np.exp(-gama*np.sum((train-test_one)*(train-test_one),axis=1)))
    result=sign(np.dot(beta,np.exp(-gama*np.sum((train-test_one)*(train-test_one),
                                                axis=1))))
    #print (result)
    return (result)
```

```python
for te in test:
    print (te)
```

```python
Gama=[32,2,0.125]
Lambda=[0.001,1,1000]
Ein_final=[]
Eout_final=[]
for gama in Gama:
    for lamda in Lambda:
        print (gama)
        print (lamda)
#         print ("K:")
        K=getK(train,gama)
#         print (K)
#         print (K.shape)
        beta=np.dot(np.linalg.inv(lamda*np.eye(len(train))+K),train.iloc[:,-1])
#         print ("beta:")
#         print (beta)
#         print (beta.shape)
        result_train=sign(np.dot(beta.reshape(1,len(beta)),K))
#         print ("train and result:")
#         print (pd.DataFrame(result_train.reshape(len(result_train[0]),1))[0])
#         print (train.iloc[:,-1])
        Ein=np.sum(pd.DataFrame(result_train.reshape(len(result_train[0]),1))[0]!=train.iloc[:,-1])/len(train)
        
        result_test=[predict(beta.reshape(1,len(beta))[0],train.iloc[:,0:-1],test.iloc[i,0:-1]) for i in range(len(test))]
#         print ("test and result:")
#         print (pd.DataFrame(result_test)[0])
#         print (test.iloc[:,-1].reset_index(drop=True))
        Eout=np.sum(pd.DataFrame(result_test)[0]!=test.iloc[:,-1].reset_index(drop=True))/len(test)
        print ("final result:")
        print (Ein)
        print (Eout)
        print ("")
        Ein_final.append(Ein)
        Eout_final.append(Eout)
        #break
    #break
print (Ein_final)
print (Eout_final)
```
